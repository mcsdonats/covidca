```python
estados = df.loc[df.place_type == 'state', ['state', 'confirmed', 'deaths', 'is_last', 'death_rate', 'estimated_population'] ]
estados.head()

estados.isnull().sum()

ultimas = estados.loc[estados.is_last == True, ['state', 'confirmed', 'deaths', 'death_rate', 'estimated_population']]
```

```python
ultimas
```

```python
ultimas['log_confirmed'] = np.log(ultimas['confirmed'])
ultimas ['log_deaths'] = np.log(ultimas['deaths'])
ultimas ['letality'] = ultimas['deaths'] / ultimas ['estimated_population']
```

```python
ultimas.head()
```

```python
import json
br_estados = "br_states.json"
geo_json_data = json.load(open(br_estados))

mapa.choropleth(
    geo_data = geo_json_data,
    name = "Letalidade",
    data = ultimas,
    columns = ['state', 'letality' ],
    key_on = 'feature.id',
    fill_color = 'Reds',
    fill_opacity = 0.6,
    line_color = "white",
    line_opacity = 0.6,
    show = True,
    legend_name = "Letalidade por Estado"
)
```
